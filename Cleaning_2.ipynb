### Interpreting Electric Retail Services
- PURCHASED: Total energy purchased by the utility (MWh) -- higher purcchases indicates larger, more established grid, lower costs
- Retail MWh: Energy distributed to end users -- higher retail signals a larger service area, potential reliability for high demand consumers
- Toal MWh: includes all energy generated and purchases, overall capacity of utility


High Cost Indicators: 
- Total MWh < 500,000 MWh -- smaller utilties may charge more

Low Cost Indicators:
- Total MWh >1,000,000 MWh -- robust network, support energy-intensive consumers at low costs
- High Retail relative to Total indicates efficient energy distribution and utilization


### For Evaluation we need to create new columns for the cost and reliability indicators:

1. Create Retail_MWh/Total_MWh Ratio: The larger the number the more efficient

2. Create Total_MWh Rank: 
- High = >1,000,000 MWh
- Moderate = 300,000 MWh > X > 1,000,000 MWh
- Low = < 300,000 MWh

3. Create Retail_MWh Rank:
- High = >1,000,000 MWh
- Moderate = 300,000 MWh > X > 1,000,000 MWh
- Low = < 300,000 MWh

```python
# Create Retail_MWh/Total_MWh Ratio column
retail_costs_df['Retail/Total MWh'] = retail_costs_df['RETAIL_MWH'] / retail_costs_df['TOTAL_MWH']
retail_costs_df.head()
```

```python
retail_costs_df.corr(numeric_only=True)
```

```python
retail_costs_df = retail_costs_df[['NAME', 'CITY', 'STATE', 'TOTAL_MWH', 'PURCHASED']]
```

```python
# Create Total_MWh Rank column
import numpy as np
conditions = [
    retail_costs_df['TOTAL_MWH'] > 1000000,
    (retail_costs_df['TOTAL_MWH'] >= 300000) & (retail_costs_df['TOTAL_MWH'] <= 1000000),
    retail_costs_df['TOTAL_MWH'] < 300000
]
choices = ['High', 'Moderate', 'Low']

retail_costs_df['Total_MWh_Rank'] = np.select(conditions, choices, default='Unknown')

# Verify the results
print(retail_costs_df['Total_MWh_Rank'].value_counts())
```

```python
# Create Purchased Rank column
conditions = [
    retail_costs_df['PURCHASED'] > 1000000,
    (retail_costs_df['PURCHASED'] >= 300000) & (retail_costs_df['PURCHASED'] <= 1000000),
    retail_costs_df['PURCHASED'] < 300000
]
choices = ['High', 'Moderate', 'Low']

retail_costs_df['Purchased_Rank'] = np.select(conditions, choices, default='Unknown')
retail_costs_df['Purchased_Rank'].value_counts()
```
