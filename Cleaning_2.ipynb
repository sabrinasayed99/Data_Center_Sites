# Loading the Electric Retail Service Territories dataset

This dataset represents electric power retail service territories. These are areas serviced by electric power utilities responsible for the retail sale of electric power to local customers, whether residential, industrial, or commercial.

We'll use the WINTR_PEAK column to drop rows with missing values. Where values are missing in WINTR Peak there also values missing in most other columns, these won't be helpful to our model.

```python
# Drop rows with -999999.0 in WINTR_PEAK
retail_costs_df = retail_costs_df[retail_costs_df['WINTR_PEAK'] != -999999.0]
retail_costs_df.info()
```

Remove rows with Canadian data

```python
retail_costs_df = retail_costs_df[retail_costs_df['COUNTRY'] == 'USA']
retail_costs_df
```

```python
retail_costs_df = retail_costs_df.drop(columns=['COUNTRY', 'ADDRESS', 'CUSTOMERS'])
retail_costs_df.info()
```

```python
retail_costs_df = retail_costs_df.drop(columns=['YEAR'])
retail_costs_df.info()
```

```python
# Drop summer and winter cap columns
retail_costs_df = retail_costs_df.drop(columns=['SUMMER_CAP', 'WINTER_CAP'])
retail_costs_df.info()
```
